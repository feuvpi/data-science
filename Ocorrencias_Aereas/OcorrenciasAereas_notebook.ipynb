- Extract data using Pandas

```python
import pandas as pd
import pandera as pa # bibilioteca utilizado para validar a extração de dados 
```

```python
## fazer carregamento do arquivo csv. Pandas ira retonar um Dataframe
df = pd.read_csv("ocorrencia3.csv", sep=",", infer_datetime_format=True, parse_dates=['ocorrencia_dia'], dayfirst=True)
df.head(10)
```

```python
# construindo um schema para o nosso dataframe
schema = pa.DataFrameSchema(
    columns = {
        "codigo": pa.Column(pa.Int, required=False),
        "codigo_ocorrencia":pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2)),
        "ocorrencia_aerodromo": pa.Column(pa.String),
        "ocorrencia_dia": pa.Column(pa.DateTime),
        # utilizando expressão regular (reg expressions) para validar dados da coluna ocorrencia_hora:
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes": pa.Column(pa.Int)
    }
)
```

```python
# realizando validação do dataframe com base no schema definido anteriormente
schema.validate(df)
```
