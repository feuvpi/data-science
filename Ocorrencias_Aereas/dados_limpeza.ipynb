## LIMPEZA DE DADOS

```python
## atribuir a celula 'ocorrencia_aerodromo' que possui valores '****' valor não disponivel
df.loc[df.ocorrencia_aerodromo == '****', ['ocorrencia_aerodromo']] = pd.NA
df.head()
```

```python
## substituir valores\
df.replace(['***', '###!', '####', '****', '*****', 'NULL'], pd.NA, inplace=True)
df.head()
```

```python
#verificar se a celula tem valor disponivel ou não
df.isna()
```
